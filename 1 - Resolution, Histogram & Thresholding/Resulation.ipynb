```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Load the grayscale image of size 512x512

```python


# Load the original image and resize it to 512x512
original_image = cv2.imread("./flower.jpg", cv2.IMREAD_GRAYSCALE)
original_image = cv2.resize(original_image, (512, 512))

# Create a copy for manipulation
decreased_image = original_image.copy()

```

```python
# Plot the images
plt.figure(figsize=(13, 13))

for k in range(1, 5):
    plt.subplot(2, 2, k)
    plt.imshow(decreased_image, cmap='gray')
    height, width = decreased_image.shape
    plt.title(f"{height}x{width}")

    # Decrease spatial resolution by half build in function
    decreased_image = cv2.resize(decreased_image, (0, 0), fx=0.5, fy=0.5)

plt.show()
```
