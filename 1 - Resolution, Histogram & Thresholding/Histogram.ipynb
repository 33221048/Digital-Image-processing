```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the grayscale image
image_path = "./bloom2.jpg"  # Replace with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Calculate and plot the histogram
hist, bins = np.histogram(image.flatten(), 256, [0, 256])
plt.plot(hist, color='black')
plt.title('Histogram of the Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Perform single threshold segmentation
threshold_value = 128  # Adjust this threshold value as needed
_, segmented_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)
```

```python
# Display the original image and segmented image side by side
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(segmented_image, cmap='gray')
plt.title(f'Segmented Image (Threshold={threshold_value})')

plt.show()
```
